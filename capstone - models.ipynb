```python
#model visualization - diminished model
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot

# model ------------------------------------------------------
model = Sequential()
model.add(Conv2D(1, kernel_size=(24, 24),
                 padding='same',
                 input_shape=(240,240,3),
                 activation='relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))#output size = 120 * 120
model.add(Conv2D(1, kernel_size=(12, 12),
                 padding='same',
                 activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))#output size = 60 * 60
model.add(Conv2D(1, kernel_size=(6, 6),
                 padding='same',
                 activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))#output size = 30 * 30
model.add(Flatten())
model.add(Dense(1 * 900, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
# visualization ------------------------------------------------------

%matplotlib inline
SVG(model_to_dot(model, show_shapes=True).create(prog='dot', format='svg'))
```

```python
#model visualization - first model
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers.convolutional import Conv2D
from keras.layers.convolutional import MaxPooling2D
from IPython.display import SVG
from keras.utils.vis_utils import model_to_dot

# model ------------------------------------------------------
model = Sequential()
model.add(Conv2D(32, kernel_size=(24, 24), # output = (240 * 240) * 3 * 32 / kernel_size * channel * kernel_num
                 padding='same',
                 input_shape=(240,240,3),
                 activation='relu'))
model.add(Conv2D(64, kernel_size=(24, 24), # output = (240 * 240) * 3 * 100
                 padding='same',
                 activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))# output = (120 * 120) * 100  = 1440000
model.add(Flatten())
model.add(Dense(128, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
# visualization ------------------------------------------------------
%matplotlib inline
SVG(model_to_dot(model, show_shapes=True).create(prog='dot', format='svg'))
```
